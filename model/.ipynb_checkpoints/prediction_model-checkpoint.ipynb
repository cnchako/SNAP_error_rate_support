```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

# load the data
df08 = pd.read_stata("qc_pub_fy2008.dta").assign(year=2008)
df09 = pd.read_stata("qc_pub_fy2009.dta").assign(year=2009)
df10 = pd.read_stata("qc_pub_fy2010.dta").assign(year=2010)
df11 = pd.read_stata("qc_pub_fy2011.dta").assign(year=2011)
df12 = pd.read_stata("qc_pub_fy2012.dta").assign(year=2012)
df13 = pd.read_stata("qc_pub_fy2013.dta").assign(year=2013)
df14 = pd.read_stata("qc_pub_fy2014.dta").assign(year=2014)
df15 = pd.read_stata("qc_pub_fy2015.dta").assign(year=2015)
df16 = pd.read_stata("qc_pub_fy2016.dta").assign(year=2016)
df17 = pd.read_stata("qc_pub_fy2017.dta").assign(year=2017)
df18 = pd.read_stata("qc_pub_fy2018.dta").assign(year=2018)
df19 = pd.read_stata("qc_pub_fy2019.dta").assign(year=2019)
df20 = pd.read_stata("qc_pub_fy2020.dta").assign(year=2020)
df21 = pd.read_stata("qc_pub_fy2021.dta").assign(year=2021)
df22 = pd.read_stata("qc_pub_fy2022.dta").assign(year=2022)
df23 = pd.read_stata("qc_pub_fy2023.dta").assign(year=2023)
```

```python
# defing a clean function 

def clean_columns(df):
    df.columns = (
        df.columns
        .str.strip()                      
        .str.lower()                     
        .str.replace(r'[\s\.-]+', '_', regex=True)  
        .str.replace(r'__+', '_', regex=True)        
    )
    return df
```

```python
# clean all the df for each year
df08 = clean_columns(df08)
df09 = clean_columns(df09)
df10 = clean_columns(df10)
df11 = clean_columns(df11)
df12 = clean_columns(df12)
df13 = clean_columns(df13)
df14 = clean_columns(df14)
df15 = clean_columns(df15)
df16 = clean_columns(df16)
df17 = clean_columns(df17)
df18 = clean_columns(df18)
df19 = clean_columns(df19)
df20 = clean_columns(df20)
df21 = clean_columns(df21)
df22 = clean_columns(df22)
df23 = clean_columns(df23)
```

```python
# concat all the data
df_all = pd.concat([
    df08, df09, df10, df11, df12, df13, df14, df15, 
    df16, df17, df18, df19, df20, df21, df22, df23
], ignore_index=True, join="outer")

# check the shape
print(df_all.shape)
print(df_all["year"].value_counts().sort_index())
```

```python
# calculate the national error rate each year
error_summary = (
    df_all.groupby("year")["error_flag"]
    .agg(["count", "sum"])
    .rename(columns={"count": "Total_Cases", "sum": "Error_Cases"})
)

error_summary["Error_Rate"] = error_summary["Error_Cases"] / error_summary["Total_Cases"]

error_summary = error_summary.round(4)  # round to 4 decimal places
print(error_summary)

```

```python
# Calculate the weighted national error rate each year
# Using FYWGT as case weight (full-year weight from QC codebook)

error_summary_wt = (
    df_all.groupby("year")
    .apply(lambda x: pd.Series({
        "weighted_total": x["fywgt"].sum(),
        "weighted_error": (x["error_flag"] * x["fywgt"]).sum(),
        "total_cases": len(x),
        "error_cases": x["error_flag"].sum()
    }))
)

error_summary_wt["weighted_error_rate"] = (
    error_summary_wt["weighted_error"] / error_summary_wt["weighted_total"]
)
error_summary_wt["unweighted_error_rate"] = (
    error_summary_wt["error_cases"] / error_summary_wt["total_cases"]
)

error_summary_wt = error_summary_wt.round(4)

print(error_summary_wt)
```

```python
# 设置风格
sns.set(style="whitegrid")

# 绘制折线图
plt.figure(figsize=(10, 6))
sns.lineplot(data=error_summary_wt, x='year', y='weighted_error_rate', marker='o', color='steelblue')

# 添加标题和标签
plt.title("National SNAP Error Rate by Year (2008–2023)", fontsize=14)
# plt.xlabel("Year", fontsize=12)
plt.ylabel("Error Rate", fontsize=12)

# 美化 y 轴百分比显示
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.1%}'))

# 添加数据点注释（可选）
for i in range(len(error_summary)):
    plt.text(
        # error_summary["year"].iloc[i],
        error_summary["Error_Rate"].iloc[i] + 0.002,
        f'{error_summary["Error_Rate"].iloc[i]:.2%}',
        ha='center', fontsize=9
    )

plt.tight_layout()
plt.show()
```

***Model***

```python
# Split by year
train_df = df_all[df_model["year"].between(2012, 2017)]
valid_df = df_all[df_model["year"].between(2018, 2019)]

print(f"Train set: {train_df.shape[0]} rows")
print(f"Validation set: {valid_df.shape[0]} rows")
```

```python
# In order to avoid multicollinearity, we just choose the representative variables from the four categories. 
feature_cols = [
    # household features
    "composition", "ctprhh",
    # policy/admin
    "expedser", "rep_sys",
    # "certmth",
    # complexity
    "fstotde2",
    # state/year fixed effects
    "state",
    # family and personal features
    'age1', 'sex1', 'raceth1', 'yrsed1',
    'empsta1', 'wrkreg1', 'fsasset', 'fsvehast', 'abwdst1',
    'ssi_cap', 'fsafil1', 'authrep', 'lastcert'
]
target_col = "error_flag"
```

```python
# ============================================
# Feature preprocessing and model training
# ============================================

# Drop rows with missing target
train = train_df.dropna(subset=[target_col])
valid = valid_df.dropna(subset=[target_col])

# Select feature columns
X_train = train[feature_cols]
y_train = train[target_col]
X_valid = valid[feature_cols]
y_valid = valid[target_col]

# Define variable type groups
cont_vars = ["ctprhh", "age1", "lastcert"]  # continuous variables
binary_vars = ["authrep", "fsasset", "fsvehast", "ssi_cap", "fsafil1", "abwdst1"]  # binary indicators
cat_vars = [
    "composition", "sex1", 
    # "raceth1", 
    "yrsed1",
    "empsta1", "wrkreg1",  # categorical personal features
    "rep_sys", "expedser", 
    # "certmth", 
    "state"  # policy and fixed effects
]

# Convert categorical variables to dummy variables
X_train = pd.get_dummies(X_train, columns=cat_vars, drop_first=True)
X_valid = pd.get_dummies(X_valid, columns=cat_vars, drop_first=True)

# Align dummy columns between train and validation sets
X_valid = X_valid.reindex(columns=X_train.columns, fill_value=0)

# Fill missing numeric values with 0 (or median if preferred)
X_train[cont_vars + binary_vars] = X_train[cont_vars + binary_vars].fillna(0)
X_valid[cont_vars + binary_vars] = X_valid[cont_vars + binary_vars].fillna(0)
```

```python
# Ensure absolutely no missing values before fitting the model
X_train = X_train.fillna(0)
X_valid = X_valid.fillna(0)

# Build the model
logit = LogisticRegression(max_iter = 1000, penalty = 'l1', solver = 'liblinear')
logit.fit(X_train, y_train)
```

```python
# predict the validation set
y_pred = logit.predict(X_valid)
y_prob = logit.predict_proba(X_valid)[:, 1]
```

```python
# model evaluation 
print("AUC:", roc_auc_score(y_valid, y_prob))
print(classification_report(y_valid, y_pred))
```

```python
# principal feature analysis
importance = pd.DataFrame({
    "feature": X_train.columns,
    "coef": logit.coef_[0]
}).sort_values("coef", key=abs, ascending=False)
```

**Prediction**

```python
# ============================================
# Prediction on test set (2020–2023)
# ============================================

# 1. Select test data by year
test_df = df_all[(df_all["year"] >= 2020) & (df_all["year"] <= 2023)]

# 2. Drop rows without target
test_df = test_df.dropna(subset=[target_col])

# 3. Select features and target
X_test = test_df[feature_cols]
y_test = test_df[target_col]

# 4. Define the same variable type groups as training
cont_vars = ["ctprhh", "age1", "lastcert"]  # continuous
binary_vars = ["authrep", "fsasset", "fsvehast", "ssi_cap", "fsafil1", "abwdst1"]  # binary
cat_vars = [
    "composition", "sex1", "raceth1", "yrsed1",
    "empsta1", "wrkreg1",  # categorical personal features
    "rep_sys", "expedser", "certmth", "state"  # policy and fixed effects
]

# 5. Dummy encoding for categorical variables
X_test = pd.get_dummies(X_test, columns=cat_vars, drop_first=True)

# 6. Align columns to match the training dataset
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

# 7. Fill missing numeric values
X_test[cont_vars + binary_vars] = X_test[cont_vars + binary_vars].fillna(0)

# 8. Ensure no missing values before prediction
X_test = X_test.fillna(0)
```

```python
# ============================================
# Use trained logistic model for prediction
# ============================================

y_pred_test = logit.predict(X_test)
y_prob_test = logit.predict_proba(X_test)[:, 1]

# ============================================
# Evaluate prediction performance
# ============================================
from sklearn.metrics import roc_auc_score, classification_report

auc = roc_auc_score(y_test, y_prob_test)
print(f"AUC (2020–2023): {auc:.4f}\n")
print(classification_report(y_test, y_pred_test, digits=2))
```

**Interpret the coefficients and marginal effects**

```python
# get the coefs and columns
coef_df = pd.DataFrame({
    "variable": X_train.columns,
    "coef": logit.coef_[0]
})
# filter all the variables start with "state_"
# coef_filtered = coef_df[~coef_df["variable"].str.startswith("state_")]

# order by the impact
coef_filtered = coef_filtered.sort_values("coef", key=abs, ascending=False)

# calculate the odds ratio
coef_filtered["odds_ratio"] = np.exp(coef_filtered["coef"])

# print the top 20 most influential variables
print(coef_filtered)
```

```python
coef_df["variable_group"] = coef_df["variable"].str.extract(r"([a-zA-Z_]+)")
group_summary = coef_df.groupby("variable_group")["coef"].agg(["mean","sum","count"]).sort_values("mean", ascending=False)
print(group_summary)
```

**State Level Function**

```python
# build a map from FIPS code to state dummy 
state_map = {
    1: "Alabama", 2: "Alaska", 4: "Arizona", 5: "Arkansas", 6: "California",
    8: "Colorado", 9: "Connecticut", 10: "Delaware", 11: "District_of_Columbia",
    12: "Florida", 13: "Georgia", 15: "Hawaii", 16: "Idaho", 17: "Illinois",
    18: "Indiana", 19: "Iowa", 20: "Kansas", 21: "Kentucky", 22: "Louisiana",
    23: "Maine", 24: "Maryland", 25: "Massachusetts", 26: "Michigan",
    27: "Minnesota", 28: "Mississippi", 29: "Missouri", 30: "Montana",
    31: "Nebraska", 32: "Nevada", 33: "New_Hampshire", 34: "New_Jersey",
    35: "New_Mexico", 36: "New_York", 37: "North_Carolina", 38: "North_Dakota",
    39: "Ohio", 40: "Oklahoma", 41: "Oregon", 42: "Pennsylvania",
    44: "Rhode_Island", 45: "South_Carolina", 46: "South_Dakota",
    47: "Tennessee", 48: "Texas", 49: "Utah", 50: "Vermont", 51: "Virginia",
    53: "Washington", 54: "West_Virginia", 55: "Wisconsin", 56: "Wyoming",
    66: "Guam", 72: "Puerto_Rico", 78: "Virgin_Islands"
}

# state names → dummy column names
state_to_dummy = {v.lower(): f"state_{k}" for k, v in state_map.items()}
```

```python
def show_state_profile(state_name):
    state_name = state_name.lower().replace(" ", "_")
    dummy_var = state_to_dummy.get(state_name)

    if dummy_var is None or dummy_var not in coef_df["variable"].values:
        print(f"No state coefficient found for '{state_name}'.")
        return None

    state_coef = coef_df.loc[coef_df["variable"] == dummy_var, "coef"].values[0]
    state_or = np.exp(state_coef)

    global_coefs = coef_df[~coef_df["variable"].str.startswith("state_")].copy()
    global_coefs["odds_ratio"] = np.exp(global_coefs["coef"])

    global_coefs["adjusted_coef"] = global_coefs["coef"] + state_coef
    global_coefs["adjusted_odds_ratio"] = np.exp(global_coefs["adjusted_coef"])

    print(f"\n===== State profile: {state_name.title()} =====")
    print(f"State fixed effect (coef): {state_coef:.3f}")
    print(f"Implied odds ratio: {state_or:.3f}\n")

    print("Top adjusted predictors (including state effect):")
    display(global_coefs.sort_values("adjusted_coef", key=abs, ascending=False).head(10)[
        ["variable", "coef", "odds_ratio", "adjusted_coef", "adjusted_odds_ratio"]
    ])
```
